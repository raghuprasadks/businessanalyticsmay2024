```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
topeconomydict ={"Country":["USA","China","Germany","Japan","India"],
                 "GDP":[27,18,4.7,4.2,3.1],
                 "PerCapita":[85,13,56,34,2]}
top5economy = pd.DataFrame(topeconomydict)
```

```python
top5economy
```

```python
print("line chart")
plt.plot(top5economy.Country,top5economy.GDP)
plt.xlabel("Country")
plt.ylabel("GDP")
plt.title("GDP of Top 5 economies")
```

```python
print("scatter plot")
plt.scatter(top5economy.Country,top5economy.GDP)
plt.xlabel("Country")
plt.ylabel("GDP")
plt.title("GDP of Top 5 economies")
```

```python
print("bar chart")
plt.bar(top5economy.Country,top5economy.GDP)
plt.xlabel("Country")
plt.ylabel("GDP")
plt.title("GDP of Top 5 economies")
```

```python
print("piechart chart")
plt.pie(top5economy.GDP,labels = top5economy.Country)
plt.xlabel("Country")
plt.ylabel("GDP")
plt.title("GDP of Top 5 economies")
```
