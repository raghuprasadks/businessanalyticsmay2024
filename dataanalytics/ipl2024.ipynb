```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#2. Create a ipl dataframe
ipl=pd.read_excel("IPL 2024.xlsx")
```

```python
#3. Display data of ipl 2024
ipl
```

```python
"""
4. Display data for following columns
Team, Matches,Points
"""
ipl[["TEAM","MATCHES","POINTS"]]
```

```python
#5. Display top 4 teams based on the points in descending order
ipl.head(4)
```

```python
#alternative way to sort and display
#ipl.POINTS.sort_values(ascending=False)
ipl.sort_values(by="POINTS",ascending=False).head(4)
```

```python
# 6. Display teams who's points is more than 12
ipl.loc[ipl.POINTS > 12,:]
```

```python
#7. Display sum of points of all teams
ipl.POINTS.sum()
```

```python
#8. Display teams whose winning percentage is more than 60 %
ipl.loc[ipl.WINS/ipl.MATCHES>0.6,:]
```

```python
#9. Draw bar chart of team vs points
plt.bar(ipl.TEAM,ipl.POINTS)
plt.title("Team Vs Points")
plt.xlabel("Team")
plt.ylabel("Points")
```

```python
#10. Draw pie chart of team vs winning percentage
ipl["winningpercentage"]=(ipl.WINS/ipl.MATCHES)*100
ipl["winningpercentage"]
plt.pie(ipl.winningpercentage,labels=ipl.TEAM)
plt.legend()
plt.title("Team vs Winning percentage")
```
